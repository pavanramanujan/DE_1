```python
import pandas as pd
```

```python
df = pd.read_csv("Downloads/netflix_titles.csv").fillna(0)
print(df.shape)
df.head(3)
```

```python
#1.Convert dates into a proper date-time format of your choice
df['date_added'] = pd.to_datetime(df['date_added'], errors='ignore').dt.strftime('%Y-%m-%d')
df.head(3)
```

```python
#2.Finding seasons count from durations columns
print("Total No.of rows in dataframe: ", df.shape[0])
print("No of rows contains season word: ", df[(df.duration.str.contains("Season", na=False))].shape[0])
print("No of rows contains seasons word: ", df[(df.duration.str.contains("Seasons", na=False))].shape[0])
```

```python
#5. how many movies were released each year
df[df.type == 'Movie'].groupby('release_year').agg({'title':'count'}).reset_index().sort_values('title', ascending = False).rename(columns = {'title':'Count'}).head(5)
```

```python
#Which month is the best month to release it`
df['month'] = pd.DatetimeIndex(df['date_added']).strftime("%B")
df.groupby('month').agg({'title':'count'}).reset_index().sort_values('title', ascending = False).rename(columns = {'title':'Count'}).head(5)
```
